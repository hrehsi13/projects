## 1. Classification model finding region

```python
x =data.drop(['region'], axis =1)
y =data['region']
```

```python
x.skew().sort_values(ascending=False)
```

```python
x_new =power_transform(x)
x = pd.DataFrame(x_new, columns= x.columns)
x.skew().sort_values(ascending=False)
```

```python
plt.figure(figsize=(20,20) )
plotnumber=1

for column in x:
    if plotnumber<=11:
        ax=plt.subplot(3,4,plotnumber)
        sns.distplot(x[column], color='red')
        plt.xlabel(column,fontsize=20)
        

    plotnumber+=1
plt.tight_layout()
```

```python
scaler = StandardScaler()
scaler.fit_transform(x)
```

```python
maxAccu=0
maxRS=0

for i in range(1,200):
    x_train,x_test,y_train,y_test =train_test_split(x,y ,test_size= 0.25, random_state =i)
    rfc =RandomForestClassifier()
    rfc.fit(x_train,y_train)
    y_pred =rfc.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    print('accuracy',acc,'random state',i)
    
    if acc > maxAccu:
        maxAccu = acc
        maxRS = i
        print('accuracy', maxAccu, 'random state',i)
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y ,test_size= 0.25, random_state =62)
```

```python
lr =LogisticRegression()
lr.fit(x_train,y_train)
y_pred =lr.predict(x_test)
```

```python
print('Accuracy score   :',accuracy_score(y_test,y_pred))
print("\n",'Confusion Matrix :',confusion_matrix(y_test,y_pred))
print("\n",classification_report(y_test,y_pred))
```

```python
scr = cross_val_score(lr,x,y, cv=10)
print("Cross Validation score",scr.mean())
```

```python
# Random Forest Classifier
rf =RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred =rf.predict(x_test)
```

```python
print('Accuracy score   :',accuracy_score(y_test,y_pred))
print("\n",'Confusion Matrix :',confusion_matrix(y_test,y_pred))
print("\n",classification_report(y_test,y_pred))
```

```python
scr = cross_val_score(rf,x,y, cv=10)
print("Cross Validation score",scr.mean())
```

```python
# Decision Tree Classifier
dtc =DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred =dtc.predict(x_test)
```

```python
print('Accuracy score   :',accuracy_score(y_test,y_pred))
print("\n",'Confusion Matrix :',confusion_matrix(y_test,y_pred))
print("\n",classification_report(y_test,y_pred))
```

```python
scr = cross_val_score(dtc,x,y, cv=8)
print("Cross Validation score",scr.mean())
```

```python
# KNN Classifier
knn =KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred =knn.predict(x_test)
```

```python
print('Accuracy score   :',accuracy_score(y_test,y_pred))
print("\n",'Confusion Matrix :',confusion_matrix(y_test,y_pred))
print("\n",classification_report(y_test,y_pred))
```

```python
scr = cross_val_score(knn,x,y, cv=8)
print("Cross Validation score",scr.mean())
```

## 2. Regression model finding Average Price

```python
x =data.drop(['AveragePrice'], axis=1)
y =data['AveragePrice']
```

```python
plt.figure(figsize = (15,15))
plotnumber = 1
for col in data.columns:
    if plotnumber <= 11:
        ax = plt.subplot(4,3, plotnumber)
        sns.distplot(data[col] ,color= 'b')
        plt.xlabel(col, fontsize =10)

    plotnumber += 1

plt.tight_layout()
plt.show()
```

```python
x.skew().sort_values(ascending=False)
```

```python
x_new =power_transform(x)
x = pd.DataFrame(x_new, columns= x.columns)
x.skew().sort_values(ascending=False)
```

```python
scaler = StandardScaler()
scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y ,test_size= 0.2, random_state= 62)
```

```python
# Linear regression

lr = LinearRegression()

lr.fit(x_train, y_train)
y_predict = lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```

```python
print(" MAE  :",metrics.mean_absolute_error(y_test,y_predict))
print(" MSE  :",metrics.mean_squared_error(y_test,y_predict))
print(" RMSE :",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
metrics.r2_score(y_test,y_predict)
```

```python
# RandomForest

rf = RandomForestRegressor()

rf.fit(x_train, y_train)
y_predict = rf.predict(x_test)
```

```python
rf.score(x_test,y_test)
```

```python
print(" MAE  :",metrics.mean_absolute_error(y_test,y_predict))
print(" MSE  :",metrics.mean_squared_error(y_test,y_predict))
print(" RMSE :",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```

```python
metrics.r2_score(y_test,y_predict)
```
